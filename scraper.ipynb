```python
#Basic Selenium Setup (Chrome)---this opens the browser automatically---

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

driver.get("https://internshala.com/internships")
time.sleep(5)
```

```python
#Extract Job Data -- this collects structured data

jobs = driver.find_elements(By.CLASS_NAME, "individual_internship")

job_data = []

for job in jobs:
    try:
        title = job.find_element(By.TAG_NAME, "h3").text
        company = job.find_element(By.CLASS_NAME, "company_name").text
        location = job.find_element(By.CLASS_NAME, "locations").text
        
        job_data.append({
            "Job Title": title,
            "Company": company,
            "Location": location
        })
    except:
        continue
```

```python
#Handle Pagination ---This is how you scale the dataset

while True:
    jobs = driver.find_elements(By.CLASS_NAME, "individual_internship")
    
    for job in jobs:
        try:
            title = job.find_element(By.TAG_NAME, "h3").text
            company = job.find_element(By.CLASS_NAME, "company_name").text
            location = job.find_element(By.CLASS_NAME, "locations").text
            
            job_data.append({
                "Job Title": title,
                "Company": company,
                "Location": location
            })
        except:
            pass

    try:
        next_btn = driver.find_element(By.LINK_TEXT, "Next")
        next_btn.click()
        time.sleep(4)
    except:
        break


```

```python
#Store Data in CSV -- now we have clean dataset
import pandas as pd

df = pd.DataFrame(job_data)
df.drop_duplicates(inplace=True)
df.to_csv("job_listings.csv", index=False)

print("Total jobs scraped:", len(df))
```
